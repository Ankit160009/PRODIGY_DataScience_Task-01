```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/API_SP.POP.TOTL_DS2_en_csv_v2_85.csv'
data = pd.read_csv(file_path, skiprows=4)

# Select countries for visualization
population_2020 = data['2020'].dropna()

# Creating the histogram
plt.figure(figsize=(10, 6))
plt.hist(population_2020, bins=20, color='skyblue', edgecolor='black')
plt.title('Global Population Distribution in 2020')
plt.xlabel('Population Size')
plt.ylabel('Number of Countries')
plt.show()
```

The histogram displays the distribution of population sizes across all countries in 2020. The x-axis represents different population sizes (binned), and the y-axis shows how many countries fall into each bin. This visualization helps identify the commonality of population sizes, such as whether most countries have small, medium, or large populations.
